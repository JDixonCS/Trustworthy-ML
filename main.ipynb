```python
import os

# Parent directory containing class subdirectories
parent_directory = "C://Users//Predator//Documents//Document-Classification//classifier//MEDTEXT"  # Replace with the actual path
#output_directory = os.path.join(parent_directory, "MEDFULL")
output_directory = "C://Users//Predator//Documents//Document-Classification//classifier\MEDFULL"

# Function to combine text files in positive and negative directories
def combine_text_files(class_dir):
    positive_dir = os.path.join(class_dir, 'pos')
    negative_dir = os.path.join(class_dir, 'neg')

    positive_output_file = os.path.join(output_directory, class_dir.split(os.path.sep)[-1] + '_positive.txt')
    negative_output_file = os.path.join(output_directory, class_dir.split(os.path.sep)[-1] + '_negative.txt')

    with open(positive_output_file, 'w') as positive_output, open(negative_output_file, 'w') as negative_output:
        for root_dir, output_file in [(positive_dir, positive_output), (negative_dir, negative_output)]:
            for subdir, _, files in os.walk(root_dir):
                for filename in files:
                    with open(os.path.join(subdir, filename), 'r', encoding="utf-8", errors='ignore') as input_file:
                        output_file.write(input_file.read())

# Get a list of subdirectories in the parent directory
class_subdirectories = [os.path.join(parent_directory, d) for d in os.listdir(parent_directory) if os.path.isdir(os.path.join(parent_directory, d))]

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Loop through the class subdirectories and combine the text files
for class_dir in class_subdirectories:
    combine_text_files(class_dir)

print("Text files have been combined into 'MEDFULL' directory for each class.")
```

```python
import os

# Parent directory
parent_directory = "C://Users//Predator//Documents//Document-Classification//classifier//MEDTEXT"  # Replace with the actual path

# Labels to process
labels = ['Immune', 'Problems_in_China', 'Risk_Factors', 'Testing', 'Transmission']

# Output directory
output_directory = "C://Users//Predator//Documents//Document-Classification//classifier//MEDFULL"  # Replace with the desired output path

# Function to separate positive and negative content for each label
def separate_sentiments_by_label(parent_dir, labels, output_dir):

  for label in labels:
    label_dir = os.path.join(parent_dir, label)

    for sentiment in ['pos', 'neg']:
      sentiment_dir = os.path.join(label_dir, sentiment)
      output_file_path = os.path.join(output_dir, f"{label}_{sentiment}.txt")

      with open(output_file_path, 'w', encoding='utf-8') as output_file:
        for root, _, files in os.walk(sentiment_dir):
          for file in files:
            if file.endswith(".txt"):
              file_path = os.path.join(root, file)
              with open(file_path, 'r', encoding='utf-8') as text_file:
                text_data = text_file.read()
                output_file.write(text_data + '\n')

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Separate positive and negative content for each label
separate_sentiments_by_label(parent_directory, labels, output_directory)

print("Text files have been created for positive and negative sentiments separately for each label.")
```

```python
import os
import pandas as pd
import re

# Parent directory containing combined positive and negative text files
parent_directory = "C://Users//Predator//Documents//Document-Classification//classifier//MEDFULL"  # Replace with the actual path

# Labels to process
labels = ['Immune', 'Problems_in_China', 'Risk_Factors', 'Testing', 'Transmission']

# Output directory for CSV files
output_directory = "C://Users//Predator//Documents//Document-Classification//classifier//MEDCSV"  # Replace with the desired output path

# Function to read positive and negative text files for a label and create a DataFrame
def create_dataframe_for_label(label, parent_dir):
    positive_file = os.path.join(parent_dir, f"{label}_positive.txt")
    negative_file = os.path.join(parent_dir, f"{label}_negative.txt")

    # Preprocess negative data
    with open(negative_file, encoding="ISO-8859-1") as raw:
        lines = raw.readlines()

        # Remove /n at the end of each line
        for index, line in enumerate(lines):
            lines[index] = line.strip()

        neg_df = pd.DataFrame(columns=['sentence'])
        i = 0
        first_col = ""
        for line in lines:
            first_col = re.sub(r' \(.*', "", line)
            neg_df.loc[i] = [first_col]
            i = i + 1

    # Preprocess positive data
    with open(positive_file, encoding="ISO-8859-1") as raw1:
        lines1 = raw1.readlines()

        # Remove /n at the end of each line
        for index, line in enumerate(lines1):
            lines1[index] = line.strip()

        pos_df = pd.DataFrame(columns=['sentence'])
        i = 0
        first_col = ""
        for line in lines1:
            first_col = re.sub(r' \(.*', "", line)
            pos_df.loc[i] = [first_col]
            i = i + 1

    # Read positive and negative sentences
    with open(positive_file, 'r', encoding='utf-8', errors='ignore') as pos_file:
        positive_sentences = pos_file.readlines()

    with open(negative_file, 'r', encoding='utf-8', errors='ignore') as neg_file:
        negative_sentences = neg_file.readlines()

    # Create DataFrames for positive and negative data
    df_positive = pd.DataFrame({'sentence': pos_df['sentence'], 'label': 1})
    df_negative = pd.DataFrame({'sentence': neg_df['sentence'], 'label': 0})

    return pd.concat([df_positive, df_negative], ignore_index=True)

# Create and save a CSV file for each label
for label in labels:
    df_label = create_dataframe_for_label(label, parent_directory)
    output_file_path = os.path.join(output_directory, f"{label}_data.csv")
    df_label.to_csv(output_file_path, index=False)
    print(f"CSV file saved for {label} at: {output_file_path}")

```
